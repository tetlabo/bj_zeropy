#### 2.7.2ã€€é–¢æ•°ã‚’è‡ªä½œã™ã‚‹

é–¢æ•°ã¯defæ–‡ã§ä½œæˆã§ãã¾ã™ã€‚å¼•æ•°ã‚’å®šç¾©ã™ã‚‹ã“ã¨ã‚‚ã§ãã¾ã™ã€‚ã“ã“ã§ã¯ã€å•†å“ã®ç¨æŠœä¾¡æ ¼ã‚’ä¸ãˆã‚‹ã¨ç¨è¾¼ä¾¡æ ¼ã‚’è¨ˆç®—ã™ã‚‹ `calc_tax()` é–¢æ•°ã‚’ä½œæˆã—ã¦ã„ã¾ã™ã€‚`price` å¼•æ•°ã«ä¾¡æ ¼ã‚’ã€`tax` å¼•æ•°ã«ç¨ç‡ã‚’æŒ‡å®šã—ã¾ã™ã€‚ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆå€¤ã¨ã—ã¦1.10ã‚’æŒ‡å®šã—ã¦ã„ã‚‹ã®ã§ã€`tax` å¼•æ•°ã«ã¤ã„ã¦ã¯çœç•¥å¯èƒ½ã§ã™ã€‚

```python
def calc_tax(price, tax=1.10):
    if price is None:
        print('å•†å“ã®ä¾¡æ ¼ã‚’priceã‚ªãƒ—ã‚·ãƒ§ãƒ³ã§æŒ‡å®šã—ã¦ãã ã•ã„ï¼')
    else:
        price_with_tax = price * tax
        price_with_tax = int(price_with_tax) # ç«¯æ•°ã®åˆ‡ã‚Šæ¨ã¦
        print('ç¨æŠœä¾¡æ ¼: ' + str(price) + 'å††')
        print('ç¨è¾¼ä¾¡æ ¼: ' + str(price_with_tax) + 'å††')
```

```python
calc_tax(350)
```

```python
# ç¨ç‡ã‚’æŒ‡å®šã™ã‚‹
calc_tax(price=500, tax=1.08)
```

å®Ÿéš›ã«æ¥­å‹™ã§å½¹ç«‹ã¤é–¢æ•°ã‚’ä½œæˆã™ã‚‹å ´åˆã€å¼•æ•°ã¨ã—ã¦æ•°å€¤ã‚’æœŸå¾…ã—ã¦ã„ã‚‹ã¨ã“ã‚ã«æ–‡å­—åˆ—ãŒä¸ãˆã‚‰ã‚ŒãŸå ´åˆã€ã©ã®ã‚ˆã†ã«æ‰±ã†ã‹ãªã©ã€å…¥åŠ›å€¤ã®ãƒã‚§ãƒƒã‚¯ã‚„ã‚¨ãƒ©ãƒ¼å‡¦ç†ã‚’ã‚ˆã‚Šå³å¯†ã«å®Ÿè£…ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

#### 2.10.1ã€€ãƒ•ã‚¡ã‚¤ãƒ«ãƒ‘ã‚¹ã‚’ã¾ã¨ã‚ã¦å–å¾—ã™ã‚‹

ç‰¹å®šã®ãƒ•ã‚©ãƒ«ãƒ€ãƒ¼ (ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªãƒ¼) ä»¥ä¸‹ã«ã‚ã‚‹ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ã¾ã¨ã‚ã¦å‡¦ç†ã—ãŸã„ã“ã¨ãŒã‚ã‚Šã¾ã™ã€‚ãã®å ´åˆã€ã¾ãšå¯¾è±¡ã¨ãªã‚‹ãƒ•ã‚¡ã‚¤ãƒ«ã®ä¸€è¦§ãƒ»å ´æ‰€ (ãƒ‘ã‚¹) ã‚’å–å¾—ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚Pythonã§ã¯ã€æ¨™æº–ã®pathlibãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ä½¿ã„ã€OSã«ä¾å­˜ã—ãªã„ãƒ•ã‚¡ã‚¤ãƒ«ãƒ‘ã‚¹ã®å–ã‚Šæ‰±ã„ãŒå¯èƒ½ã§ã™ã€‚

```python
from pathlib import Path
p = Path('/content/drive/zeropy/chap02/txt') # ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªã‚’æŒ‡å®š
file_list = list(p.glob('**/*.txt')) # ãƒ•ã‚¡ã‚¤ãƒ«ä¸€è¦§ã‚’å–å¾—

[str(file) for file in file_list] # ãƒ•ã‚¡ã‚¤ãƒ«ä¸€è¦§ã‚’æ–‡å­—åˆ—ã¨ã—ã¦å‡ºåŠ›
```

ã“ã“ã§ã¯ã€ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã¨ã„ã†æ›¸å¼ã§ã€ç¹°ã‚Šè¿”ã—å‡¦ç†ã‚’ç°¡æ½”ã«è¨˜è¿°ã—ã¦ã„ã¾ã™ã€‚ç¹°ã‚Šè¿”ã—å‡¦ç†ã«ã¤ã„ã¦ã¯å¾Œè¿°ã—ã¦ã„ã¾ã™ã€‚ãƒ‡ãƒ¼ã‚¿ã¯ã€GitHubã«å…¬é–‹ã•ã‚Œã¦ã„ã‚‹ã€[æ‰€ä¿¡è¡¨æ˜æ¼”èª¬ã‚³ãƒ¼ãƒ‘ã‚¹](https://github.com/yuukimiyo/GeneralPolicySpeechOfPrimeMinisterOfJapan)ã‚’ä½¿ç”¨ã—ã¦ã„ã¾ã™ã€‚

#### 2.10.2ã€€foræ–‡ã«ã‚ˆã‚‹ç¹°ã‚Šè¿”ã—å‡¦ç†

å¯¾è±¡ã¨ãªã‚‹ãƒ•ã‚¡ã‚¤ãƒ«ã®ä¸€è¦§ã‚’å–å¾—ã—ãŸã‚‰ã€é †ã«å‡¦ç†ã‚’é©ç”¨ã—ã¦ã„ãã¾ã™ã€‚Pythonã«ã¯ã€ç¹°ã‚Šè¿”ã— (ãƒ«ãƒ¼ãƒ—) å‡¦ç†ã®ãŸã‚ã®åˆ¶å¾¡æ§‹é€ ãŒç”¨æ„ã•ã‚Œã¦ã„ã¾ã™ã€‚ã“ã“ã§ã¯ã€åºƒãä½¿ã‚ã‚Œã‚‹foræ–‡ã®ä¾‹ã‚’ç´¹ä»‹ã—ã¾ã™ã€‚

#### 2.10.3ã€€ifæ–‡ã«ã‚ˆã‚‹æ¡ä»¶åˆ†å²

å‡¦ç†å¯¾è±¡ã®ãƒ•ã‚¡ã‚¤ãƒ«ãªã©ãŒã€ã‚ã‚‹æ¡ä»¶ã«ä¸€è‡´ã™ã‚‹ã¨ãã ã‘å‡¦ç†ã‚’é©ç”¨ã—ãŸã„å ´åˆãŒã‚ã‚Šã¾ã™ã€‚ãã®ã‚ˆã†ãªå ´åˆã€ifæ–‡ã‚’ä½¿ã†ã“ã¨ã§æŸ”è»Ÿãªæ¡ä»¶åˆ†å²ã‚’æŒ‡å®šã§ãã¾ã™ã€‚ifæ–‡ã§ã¯ã€æ¡ä»¶ã«åˆè‡´ã—ãŸå ´åˆã®å‡¦ç†ã®ã¿å®šç¾©ã™ã‚‹ãƒ‘ã‚¿ãƒ¼ãƒ³ã€æ¡ä»¶ã«åˆè‡´ã—ãªã„å ´åˆã®å‡¦ç†ã‚‚å®šç¾©ã™ã‚‹ãƒ‘ã‚¿ãƒ¼ãƒ³ã€ãã—ã¦ã€è¤‡æ•°ã®æ¡ä»¶ã”ã¨ã«å‡¦ç†ã‚’å®šç¾©ã™ã‚‹ãƒ‘ã‚¿ãƒ¼ãƒ³ãŒå®Ÿè£…ã§ãã¾ã™ã€‚ä»¥ä¸‹ã§ã€ãã‚Œãã‚Œã®ä¾‹ã‚’ç¤ºã—ã¾ã™ã€‚

```python
# æ¡ä»¶ã«åˆè‡´ã—ãŸå ´åˆã®å‡¦ç†ã®ã¿å®šç¾©ã™ã‚‹
import random # æ¨™æº–ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®randomãƒ¢ã‚¸ãƒ¥ãƒ¼ãƒ«
num = random.randint(1, 20) # 1ä»¥ä¸Š20ä»¥ä¸‹ã®ç¯„å›²ã®ä¹±æ•°ã‚’ç”Ÿæˆ

if num % 2 == 0:
    print(str(num) + 'ã¯å¶æ•°ã§ã™ã€‚')
```

```python
# æ¡ä»¶ã«åˆè‡´ã—ãªã„å ´åˆã®å‡¦ç†ã‚‚å®šç¾©ã™ã‚‹ãƒ‘ã‚¿ãƒ¼ãƒ³
num = random.randint(1, 20) # 1ä»¥ä¸Š20ä»¥ä¸‹ã®ç¯„å›²ã®ä¹±æ•°ã‚’ç”Ÿæˆ

if num % 2 == 0:
    print(str(num) + 'ã¯å¶æ•°ã§ã™ã€‚')
else:
    print(str(num) + 'ã¯å¥‡æ•°ã§ã™ã€‚')
```

```python
# è¤‡æ•°ã®æ¡ä»¶ã”ã¨ã«å‡¦ç†ã‚’å®šç¾©ã™ã‚‹ãƒ‘ã‚¿ãƒ¼ãƒ³
num = random.randint(1, 20) # 1ä»¥ä¸Š20ä»¥ä¸‹ã®ç¯„å›²ã®ä¹±æ•°ã‚’ç”Ÿæˆ
if (num % 2 == 0) & (num < 10):
    print(str(num) + 'ã¯10æœªæº€ã®å¶æ•°ã§ã™')
elif (num % 2 == 0) & (num >= 10):
    print(str(num) + 'ã¯10ä»¥ä¸Šã®å¶æ•°ã§ã™')
elif (num % 2 != 0) & (num < 10):
    print(str(num) + 'ã¯10æœªæº€ã®å¥‡æ•°ã§ã™')
else:
    print(str(num) + 'ã¯10ä»¥ä¸Šã®å¥‡æ•°ã§ã™')
```
